이미지 데이터 크기 확인

```python
from PIL import Image
import os

# 이미지 파일이 있는 폴더의 경로
folder_path = '/content/drive/MyDrive/GDSC/can/'  # 경로 확인 필요

# 폴더 내의 모든 파일에 대해 반복
for filename in os.listdir(folder_path):
    # 파일의 전체 경로를 구성
    file_path = os.path.join(folder_path, filename)

    # 파일이 이미지인 경우 크기를 확인
    try:
        with Image.open(file_path) as img:
            print(f"{filename}: {img.size}")  # img.size는 (너비, 높이) 튜플을 반환합니다.
    except IOError:
        print(f"Deleting non-image or corrupt file: {filename}")
        os.remove(file_path)  # 파일 삭제
```

# 이미지 전처리 크기 변경

```python
import tensorflow as tf
from tensorflow.keras.preprocessing import image
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.image import pad_to_bounding_box
from tensorflow.image import central_crop
from tensorflow.image import resize
```

```python
#이미지 불러오기
bgd = image.load_img('/content/drive/MyDrive/GDSC/can/can_002.jpg')
bgd_vector = np.asarray(image.img_to_array(bgd))
bgd_vector = bgd_vector/255

#이미지 형태 확인
bgd_vector.shape

#이미지 확인
plt.imshow(bgd_vector)
plt.show()
```

```python
w, h = bgd.size

s = min(w, h) #둘 중에 작은 것 기준으로 자름
y = (h - s) // 2
x = (w - s) // 2

print(w, h, x, y, s)

# 좌, 위, 오른쪽, 아래 픽셀 설정
bgd = bgd.crop((x, y, x+s, y+s))
plt.imshow(np.asarray(bgd))
bgd.size
```

```python
# 이미지를 224x224 크기로 조정
bgd_resized = resize(bgd, [224, 224])

# 이미지 확인
plt.imshow(bgd_resized.numpy().astype('uint8'))
plt.show()
```
