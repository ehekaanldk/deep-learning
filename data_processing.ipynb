# 이미지 데이터 가공
- 중심 기준으로 center cropping
- 이미지 크기 224x224 통일

```python
import tensorflow as tf
from tensorflow.keras.preprocessing import image
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import os

# 이미지가 저장된 폴더의 경로
folder_path = '/content/drive/MyDrive/GDSC/can/'

# 저장할 폴더의 경로 (경로가 없으면 생성)
save_folder_path = '/content/drive/MyDrive/GDSC/can_resized/'
if not os.path.exists(save_folder_path):
    os.makedirs(save_folder_path)

# 폴더 내의 모든 이미지 파일에 대해 반복
for filename in os.listdir(folder_path):
    file_path = os.path.join(folder_path, filename)

    # 이미지 불러오기
    with Image.open(file_path) as img:
        w, h = img.size
        s = min(w, h)  # 둘 중에 작은 것을 기준으로 자름
        y = (h - s) // 2
        x = (w - s) // 2

        # 중앙에서 이미지 자르기
        img_cropped = img.crop((x, y, x+s, y+s))

        # TensorFlow를 사용하여 이미지 크기 조정
        img_cropped_array = np.asarray(img_cropped) / 255.0
        img_resized = tf.image.resize(img_cropped_array, [224, 224])
        img_resized = img_resized.numpy()

        # PIL 이미지로 변환하여 저장
        img_resized_pil = Image.fromarray((img_resized * 255).astype('uint8'))
        img_resized_pil.save(os.path.join(save_folder_path, filename))

print("Image resizing and saving completed!")
```
